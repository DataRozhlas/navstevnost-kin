```python
def uprava_sloupcu(frame):
    sloupce = frame.columns.to_list()
    slovnicek = {}
    for s in sloupce:
        slovnicek[s] = s.lower().replace('.','').replace(' ','_')
    return(slovnicek)
```

```python
df = pd.DataFrame()
poradi = [x for x in range(1,51)]
for r in rocni:
    print(r)
    df_r = pd.read_excel(os.path.join('downloads',r))
    rok = re.search('\d\d\d\d',r.replace('top50','')).group()
    if rok == '1412':
        rok = '2014'
    df_r['rok'] = rok
    slovnik = df_r.iloc[2].to_dict()
    sloupce = []
    for key, val in slovnik.items():
        if val not in sloupce:
            sloupce.append(val)
        else:
            slovnik[key] = f'{val}_celkem'
    slovnik['rok'] = 'rok'
    for key, val in slovnik.items():
        slovnik[key] = val.lower()
    df_r = df_r.rename(columns=slovnik)
    rank = df_r.columns.to_list()[0]
    df_r = df_r[df_r[rank].isin(poradi)].reset_index()
    df = pd.concat([df, df_r])
df = df.drop(['index'],axis=1)
```

## Týdenní statistiky

```python
def tydenni(soubor):
    try:
        df_t = pd.read_excel(os.path.join('downloads',soubor))
        slovnik_popisky = df_t.iloc[3].to_dict()
        slovnik_poznamky = df_t.iloc[4].to_dict()
        for key in slovnik_popisky:
            if str(slovnik_poznamky[key]) != "nan":
                slovnik_popisky[key] += f" {slovnik_poznamky[key]}"
        df_t = df_t.rename(columns=slovnik_popisky)
        # df_t = df_t[df_t['Poř'].isin(poradi)]
        rok = soubor.replace('top20','').replace('-','')[0:4]
        df_t['rok'] = rok
        tyden = soubor.replace('top20','').replace('-','').replace(rok,'').split('cz')[0]
        df_t['tyden'] = tyden
        prejmenovani = uprava_sloupcu(df_t)
        df_t = df_t.rename(columns=prejmenovani)
        df_t = df_t[df_t['poř'].isin(poradi)]
        return df_t
    except:
        return pd.DataFrame()
```
