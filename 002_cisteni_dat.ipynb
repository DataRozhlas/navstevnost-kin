```python
def uprava_sloupcu(frame):
    sloupce = frame.columns.to_list()
    slovnicek = {}
    for s in sloupce:
        slovnicek[s] = s.lower().replace('.','').replace(' ','_')
    return(slovnicek)
```

```python
df = pd.DataFrame()
poradi = [x for x in range(1,51)]
for r in rocni:
    print(r)
    df_r = pd.read_excel(os.path.join('downloads',r))
    rok = re.search('\d\d\d\d',r.replace('top50','')).group()
    if rok == '1412':
        rok = '2014'
    df_r['rok'] = rok
    slovnik = df_r.iloc[2].to_dict()
    sloupce = []
    for key, val in slovnik.items():
        if val not in sloupce:
            sloupce.append(val)
        else:
            slovnik[key] = f'{val}_celkem'
    slovnik['rok'] = 'rok'
    for key, val in slovnik.items():
        slovnik[key] = val.lower()
    df_r = df_r.rename(columns=slovnik)
    rank = df_r.columns.to_list()[0]
    df_r = df_r[df_r[rank].isin(poradi)].reset_index()
    df = pd.concat([df, df_r])
df = df.drop(['index'],axis=1)
```

```python
def tydenni(soubor):
    
    df_t = pd.read_excel(os.path.join('downloads',soubor), engine='calamine')
    try:
        vikend_od = re.search('end od:\s{0,1}\d{1,2}\.\d{1,2}\.\d\d\d\d',df_t.to_string()).group().split('od:')[1].strip()
    except:
        vikend_od = None
    try:
        tyden_od = re.search('en od:\s{0,1}\d{1,2}\.\d{1,2}\.\d\d\d\d',df_t.to_string()).group().split('od:')[1].strip()
    except:
        tyden_od = None
    try:
        datum = re.search('atum\s{0,1}=\s{0,1}\d{1,2}\.\d{1,2}\.\d\d\d\d',df_t.to_string()).group().split('od:')[1].strip()
    except:
        datum = None
    
    for i in range(0,7):
        if ("změna" in df_t.iloc[i].to_string().lower()) and ("celkem" in df_t.iloc[i].to_string().lower()):
            slovnik_popisky = df_t.iloc[i].to_dict()
        if "tržby" in df_t.iloc[i].to_string().lower():
            slovnik_poznamky = df_t.iloc[i].to_dict()
    for key in slovnik_popisky:
        if str(slovnik_poznamky[key]) != "nan":
            slovnik_popisky[key] = f"{slovnik_popisky[key]}_{slovnik_poznamky[key]}"
    df_t = df_t.rename(columns=slovnik_popisky)
    if soubor[1] == '-':
        soubor = soubor[2:]
    rok = soubor.replace('kopie-','').replace('top20','').replace('-','')[0:4]
    df_t['rok'] = rok
    tyden = soubor.replace('kopie-','').replace('top20','').replace('-','').replace(rok,'').split('cz')[0]
    df_t['týden'] = tyden
    if soubor == '48-2011.xls':
        df_t['rok'] = 2011
        df_t['týden'] = 48
    prejmenovani = uprava_sloupcu(df_t)
    df_t = df_t.rename(columns=prejmenovani)
    df_t = df_t[df_t['poř'].isin(poradi)]
    df_t = df_t.T.groupby(level=0).first().T
    
    df_t['týden_od'] = tyden_od
    df_t['víkend_od'] = vikend_od
    df_t['datum'] = datum
    
    return df_t
```
