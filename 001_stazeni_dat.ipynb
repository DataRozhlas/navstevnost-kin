## Týdenní statistiky

```python
tydny = "https://www.ufd.cz/top-20-filmu/top-20-cr?page="
```

```python
if len(xls_files) < 500:
    rozmezi = [i for i in range(0,22)]
else:
    rozmezi = [i for i in range(0,1)]
```

```python
tydenni_statistiky = []
for i in rozmezi:
    link = f"{tydny}{i}"
    print(link)
    response = requests.get(link)
    soup = BeautifulSoup(response.text, "html.parser")
    for o in soup.find_all('a'):
        if ".xls" in o['href']:
            tydenni_statistiky.append(o['href'])
```

```python
for url in tydenni_statistiky:
    if url.split('/')[-1] not in xls_files:
        print(url.split('/')[-1])
        response = requests.get(url)
        with open(os.path.join("downloads", url.split('/')[-1]), 'wb') as f:
            f.write(response.content)
```
