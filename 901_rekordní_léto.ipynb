## První graf: celková návštěvnost

```python
dosrpna = pd.Series(df[df['mesic'] <= 8].groupby('rok')['divactvo'].sum(), name="leden-srpen")
```

```python
dosrpna
```

```python
celyrok = pd.Series(df.groupby('rok')['divactvo'].sum(), name="celý rok (letošek do srpna včetně)")
```

```python
prazdniny = pd.Series(df[df['mesic'].isin([7,8])].groupby('rok')['divactvo'].sum(), name="červenec a srpen")
```

```python
prazdniny
```

```python
irozhlas_graf(sloupcovy=[prazdniny, celyrok, dosrpna], skryte=[celyrok, dosrpna], target="navstevnost-kin", kredity=['zdroj dat: UFD.cz, vizualizace: iROZHLAS.cz','https://www.irozhlas.cz/zpravy-tag/datova-zurnalistika'], titulek="Návštěvnost českých kin v létě vs. zbytcích roku", podtitulek="", barvy=['#fd7f6f','#ffb55a',"#beb9db"])
```

## Druhý graf: letošní hity

```python
filmy_stare = tydenni[tydenni['rok'] <= 2023]['film'].drop_duplicates().to_list()
```

```python
filmy_nove = tydenni[~tydenni['film'].isin(filmy_stare)].groupby('film')['celkem_diváci'].max().nlargest(9).index.to_list()
```

```python
filmy_nove
```

```python
letosni_hity = []
prejmenovani = {}
for w in range(1,53):
    prejmenovani[w] = f"{w}. týden"
for f in filmy_nove:
    letosni_hity.append(pd.Series(tydenni[tydenni['film'] == f].groupby(['týden'])['celkem_diváci'].max().reindex([w for w in range(1,53)]).rename(index=prejmenovani), name=f))
```

```python
letosni_hity
```

```python
irozhlas_graf(carovy=letosni_hity, target="letosni-hity", kredity=['zdroj dat: UFD.cz, vizualizace: iROZHLAS.cz','https://www.irozhlas.cz/zpravy-tag/datova-zurnalistika'], titulek="Kumulovaná návštěvnost nejúspěšnějších letošních filmů", skryte=letosni_hity[6:], skrytnuly=True, prvni=False, podtitulek="Za povšimnutí stojí, že pět nejnavštěvovanějších šlo do kin právě v létě. Přerušení čáry značí vypadnutí z týdenní top dvacítky.")
```
