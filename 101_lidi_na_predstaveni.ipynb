```python
import pandas as pd
```

```python
df = pd.read_csv("data/souhrnne_vysledky.csv")
```

```python
df.groupby('rok')['divactvo'].sum().plot.bar()
```

```python
df.groupby('rok')['divactvo'].sum().sort_values()
```

```python
(df.groupby('rok')['divactvo'].sum() / df.groupby('rok')['predstaveni'].sum()).plot.bar(figsize=(6,4), title="Průměrná návštěvnost filmového představení", xlabel="")
```
